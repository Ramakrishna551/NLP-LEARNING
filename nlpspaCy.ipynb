```python
#SPACY CODES 
#spaCy :spaCy is a free and open=source library for advanced Natural Language Processing 
#in python .It is written in cython language(C extension of python) which gives C like performance to python.
#It is fast and provides concise API's to access it's methods and properties 
```

```python
import spacy
nlp=spacy.load("en_core_web_sm")

doc=nlp("""Google maps street veiw tean tracks locations across the world,uploading the images to its websites,some have been deemed the scariest, or creeoiest , sites ever-including a sun-scorched field in North Corolina.""")
for token in doc:
    print(token.text)
```

```python
#Lemmatization with spacy
for lemma in doc:
    print(lemma.text,lemma.lemma_)
```

```python
#POS-(Part-of-speech) with spacy
#To print pos tags
doc=nlp("""Google maps street veiw tean tracks locations across the world,uploading the images to its websites,some have been deemed the scariest, or creeoiest , sites ever-including a sun-scorched field in North Corolina.""")
for token in doc:
    print(token.text+"---",token.pos_+"---",token.tag_)
 
```

```python
#to use visualizers for pos tags
import spacy
from spacy import displacy
nlp=spacy.load("en_core_web_sm")
doc=nlp(""" Tokyo is the largest city in the world , if the entire Tokyo metro area is included ,with a total of more than 38 million residents .Another Japanese city ,Osaka also has a very large population of almost 20.5 million .""")
d=nlp("Natural Language Processing is Fun!")
displacy.render(d,style="dep")
displacy.render(doc,style="dep")
```

```python
#Named Entity Recognition with spacy
#to print named entities
import spacy
nlp=spacy.load("en_core_web_sm")
doc=nlp(""" Tokyo is the largest city in the world , if the entire Tokyo metro area is included ,with a total of more than 38 million residents .Another Japanese city ,Osaka also has a very large population of almost 20.5 million .""")
hello=nlp("Apple is looking at buying U.K. startup for $1 billion")
for ent in doc.ents:
    print(ent.text+"---",ent.label_)
for ent in hello.ents:
    print(ent.text+"--",ent.label_)    
```

```python
#to use visualizer for NER (Named -Entity - Recognition )
import spacy 
from spacy import displacy
nlp=spacy.load("en_core_web_sm")
doc=nlp(""" Tokyo is the largest city in the world , if the entire Tokyo metro area is included ,with a total of more than 38 million residents .Another Japanese city ,Osaka also has a very large population of almost 20.5 million .""")
displacy.render(doc,style="ent")
```
