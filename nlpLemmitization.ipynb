```python
#STARTED Lemmitization
from nltk.stem import WordNetLemmatizer
lematizer = WordNetLemmatizer()
print(lematizer.lemmatize("corpora"))
print(lematizer.lemmatize("cacti"))
print(lematizer.lemmatize("lions"))
print(lematizer.lemmatize("rocks"))
print(lematizer.lemmatize("geese"))
print(lematizer.lemmatize("better",pos="a"))
print(lematizer.lemmatize("runs"))
print(lematizer.lemmatize("runs",pos="v"))
```

```python
#example 2 Lemmitization
from nltk.stem import WordNetLemmatizer
ps=WordNetLemmatizer()
words=["cats","feet","bats","corpora"]
for w in words:
    print(ps.lemmatize(w))
```

```python
#Part-of-speech tagging 
#In pos Tagging the task is to label/tag each word in a sentene with its grammatical group such as noun,pronoun 
#verb,adverb,adjective,tense and many more 
import nltk
from nltk import pos_tag
from nltk.tokenize import word_tokenize
sample=""""Google has actively been purging malware-laden apps from the play store,but the company has been having a hard time with its ,especially when some very popular apps are found to be complicit."""
tokent_text=word_tokenize(sample)
print(tokent_text)
nltk.pos_tag(tokent_text)
```

```python
#Name-Entity-Recognition with nltk
#Named-Entity-Recognition is used to identify import named entities in a text such as people ,places , Organization ,locations
#and dates ,states work of art etc.
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import state_union
sample=state_union.raw("2005-GWBush.txt")
tokenized_text=nltk.word_tokenize(sample)
print(tokenized_text)
pos_words=nltk.pos_tag(tokenized_text)
from nltk import ne_chunk
ner_words=ne_chunk(pos_words)
print(ner_words)
```
